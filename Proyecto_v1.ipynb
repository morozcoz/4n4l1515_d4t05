```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Paso 1: Agrupar por departamento y tecnolog√≠a, y sumar el √°rea total
grouped = ndf.groupby(['DEPARTAMENTO', 'TECNOLOGIA'])['TOTAL_AREA_COB'].sum().unstack()

# Paso 2: Ordenar por el √°rea total de cobertura de 4G
grouped = grouped.sort_values(by='4G', ascending=False)  # Ordena de mayor a menor

# Paso 3: Preparar los datos para el gr√°fico de barras apiladas
departamentos = grouped.index
A2g = grouped['2G']
A3g = grouped['3G']
A4g = grouped['4G']
A5g = grouped['5G']

# Paso 4: Crear el gr√°fico de barras apiladas
plt.figure(figsize=(12, 6))

# Usamos 'bottom' para apilar las barras
plt.bar(departamentos, A2g, label='2G')
plt.bar(departamentos, A3g, bottom=A2g, label='3G')
plt.bar(departamentos, A4g, bottom=A2g + A3g, label='4G')
plt.bar(departamentos, A5g, bottom=A2g + A3g + A4g, label='5G')

plt.xlabel('Departamento')
plt.ylabel('√Årea de Cobertura')
plt.title('√Årea de Cobertura por Departamento y Tecnolog√≠a (Ordenado por 4G)')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
import plotly.express as px

# Ruta al archivo .shp (mapa de Colombia)
shapefile_path = 'https://github.com/morozcoz/4n4l1515_d4t05/raw/refs/heads/main/Col_1/gadm41_COL_1.shp'

# Cargar el shapefile (mapa)
colombia = gpd.read_file(shapefile_path)

# Filtrar por campos ANNO y Trimestre
df_comb_filtrado = df_comb[(df_comb['ANNO'] == 2024) & (df_comb['TRIMESTRE'] == 2)]

# DataFrame con nombres de departamentos y sus coordenadas espec√≠ficas
data_departamentos = {
    'DEPARTAMENTO': df_dpto['Nombre'],
    'Latitud': df_dpto['Latitud'],
    'Longitud': df_dpto['Longitud']
}

df_departamentos = pd.DataFrame(data_departamentos)

# Convertir a geometr√≠a de puntos para los departamentos
geometry_departamentos = [Point(xy) for xy in zip(df_departamentos['Longitud'], df_departamentos['Latitud'])]
gdf_departamentos = gpd.GeoDataFrame(df_departamentos, geometry=geometry_departamentos)

# DataFrame con datos que se requieren marcar en el mapa
data = {'Longitud': df_comb_filtrado['Longitud'], 'Latitud': df_comb_filtrado['Latitud'], 'TECNOLOGIA': df_comb_filtrado['TECNOLOGIA']}
df_point = pd.DataFrame(data)

# Convertir a geometr√≠a de puntos para cada tecnolog√≠a
geometry = [Point(xy) for xy in zip(df_point['Longitud'], df_point['Latitud'])]
gdf = gpd.GeoDataFrame(df_point, geometry=geometry)

# Definir colores y tama√±os para cada tecnolog√≠a
colores = {'2G': 'black', '3G': '#DEFECA', '4G': 'orange', '5G': '#0099DD', 'Ninguna': 'pink'}
tama√±os = {'2G': 20, '3G': 15, '4G': 10, '5G': 5, 'Ninguna': 2}

# A√±adir colores y tama√±os a los datos, manejando valores NaN
gdf['color'] = gdf['TECNOLOGIA'].map(colores).fillna('gray')
gdf['size'] = gdf['TECNOLOGIA'].map(tama√±os).fillna(5)

# Convertir a DataFrame para plotly
gdf['Longitud'] = gdf.geometry.x
gdf['Latitud'] = gdf.geometry.y
gdf_departamentos['Longitud'] = gdf_departamentos.geometry.x
gdf_departamentos['Latitud'] = gdf_departamentos.geometry.y

# Crear el mapa interactivo con plotly
fig = px.scatter_mapbox(
    gdf,
    lat='Latitud',
    lon='Longitud',
    color='TECNOLOGIA',
    size='size',
    size_max=20,
    zoom=3.5,  # Ajuste el zoom para que incluya San Andr√©s y el continente
    mapbox_style='carto-positron',
    title='Cobertura redes m√≥viles en Colombia',
    height=800,  # Ajusta la altura del gr√°fico
    width=1200,   # Ajusta el ancho del gr√°fico
    color_discrete_map=colores  # Asignar los colores definidos
)

# A√±adir los departamentos al mapa
fig.add_scattermapbox(
    lat=gdf_departamentos['Latitud'],
    lon=gdf_departamentos['Longitud'],
    mode='markers+text',
    text=gdf_departamentos['DEPARTAMENTO'],
    textposition='top right',
    marker=dict(size=5, color='green'),
    showlegend=False
)

# Ajustar el layout para mejorar la visualizaci√≥n y asegurar que San Andr√©s est√© incluido
fig.update_layout(
    legend_title_text='Tecnolog√≠a',
    margin={'r':10,'t':50,'l':10,'b':10},  # Ajustar los m√°rgenes
    mapbox=dict(
        center=dict(lat=5.0, lon=-70.0),  # Centro del mapa para incluir Colombia y San Andr√©s
        zoom=4.8  # Ajuste el zoom para incluir San Andr√©s y el continente
    ),
        title_font=dict(size=28, family='Calibri', color='black', weight='bold')
)

# Mostrar el mapa
fig.show()
```
